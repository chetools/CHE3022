```python
import numpy as np
from plotly.subplots import make_subplots
import plotly.io as pio
pio.templates.default='plotly_dark'
from scipy.linalg import solve_banded
```

```python
nx = 50
nt = 2000

alpha = 1. 
deltax = 1.
deltat = 0.5
x = np.arange(nx)*deltax
r = alpha * deltat/deltax**2
T=np.full(nx,20.)
T[0]=50
T[-1]=80
fig = make_subplots(rows=1,cols=1)
fig.add_scatter(x=x,y=T,name='initial')
for i in range(1,nt+1):
  T[1:-1]=T[1:-1]+r*(T[2:]-2*T[1:-1] + T[:-2])
  if i % 100 == 0:
    fig.add_scatter(x=x, y=T, mode='lines', name=f'{i}')
fig.show()
```

```python
nx = 100  # number of spatial divisions (nodes)
nt = 2000 # number of time steps

alpha = 1. # thermal diffusivity of the material
deltax = 0.5 #sptial increment
deltat = 0.5 # time increment
x = np.arange(nx)*deltax # generate list of spatial positions for plotting
r = alpha * deltat/deltax**2/2 # defining r for Crank-Nicolson - notice 2 in the denominator 
```

```python
T=np.full(nx,20.) # set initial temperature to be 20C uniform
T[0]=50. # left-side temperature starts and stays at 50C
T[-1]=80. # right-side temperature
```

```python
A=np.zeros((3,nx-2)) # matrix of diagonals - 3 rows for upper, main, lower diagonals, nx-2 columns because temperatures at left and right ends are set
A[0,1:]=-r #upper diagonal
A[1,:]=1+2.*r #main diagonal
A[2,:-1]=-r #lower diagonal
b=np.zeros(nx-2) # allocate vector of rhs of set of linear equations
b=r*T[2:] + (1-2.*r) * T[1:-1] + r*T[:-2]
b[0]=b[0]+r*T[0]
b[-1]=b[-1]+r*T[-1]
fig2 = make_subplots(rows=1,cols=1)
fig2.add_scatter(x=x,y=T, name='initial')
for i in range(nt+1):
    T[1:-1]=solve_banded((1,1),A,b)
    b=r*T[2:] + (1-2.0*r) * T[1:-1] + r*T[:-2]
    b[0]=b[0]+r*T[0]
    b[-1]=b[-1]+r*T[-1]
    if i % 100 == 0:
        fig2.add_scatter(x=x,y=T, name=f'{i*deltat}s')

```
